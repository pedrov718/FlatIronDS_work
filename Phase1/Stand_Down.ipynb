```python
import pandas as pd
```

```python
df = pd.read_csv("vgsales-12-4-2019-short.csv")
```

```python
df.drop(columns = "User_Score")
```

```python
percent_action = len(df.loc[df["Genre"] == "Action"])/ len(df)
print(f"The percent of games in the action genre is {percent_action * 100:.02f}%")
```

```python
df["Genre"].value_counts(normalize = True).plot(kind = "bar")
```

```python
df.dropna(subset= ["Year", "Total_Shipped"], inplace= True)
df
```

imput missing critic score values with the average critic score

```python
df.Critic_Score.fillna(value = df["Critic_Score"].mean(), inplace = True )
df
```

```python
df.groupby('Year').count()['Total_Shipped'].plot()
```

```python
df.groupby('Year').sum()['Total_Shipped'].plot()
```

```python
df.groupby(["Year"])["Total_Shipped"].count().sort_values(ascending= False)
```
