```python
import pandas as pd
import matplotlib as plt
salary_file = "ds_salaries.csv"
```

```python
salary_df = pd.read_csv(salary_file)
```

```python
salary_df.keys()
```

```python
salary_df.sample(10)
```

```python
salary_df.drop(columns = ["Unnamed: 0"], inplace= True)
```

```python
salary_df.describe()
```

```python
salary_df.loc[salary_df["salary_in_usd"] == salary_df["salary_in_usd"].max()]
```

```python
salary_df.loc[salary_df["salary_in_usd"] == salary_df["salary_in_usd"].min()]
```

```python
average_sal_pos = salary_df.groupby("job_title")["salary_in_usd"].mean().sort_values(ascending= False)
```

```python
average_sal_pos.plot.barh(color = "red", figsize = (20,10))
```
